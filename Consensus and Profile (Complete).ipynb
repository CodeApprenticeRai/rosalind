```python
def consensus_and_profile(filename):
    #Get Input
    f = open(filename, 'r').read()
    
    #Filter for DNA String
    raw0 = re.split(">Rosalind_\d{4}",f) #[\d\n]"
    # raw1 = raw0.split("\n")
    raw1 = list(filter(lambda x: len(x) > 15, raw0))
    raw1 = [ s.replace("\n", "") for s in raw1]
 

    matrix = []
    #Perform counts
    for j in range(len(raw1[0])):
        table = {'A':0, 'C':0, 'G':0, 'T':0}
        for i in range(len(raw1)):
            table[raw1[i][j]] += 1
        matrix.append(table)

    maxes = []
    #Formatting and printing consensus string
    for d in matrix:
        max0 = 'A'
        for thing in d:
             if d[max0] < d[thing]:
                    max0 = thing
        maxes.append(max0)

    for char in maxes:
        print(char, end="")
    
    print("")
    #Formatting profile matrix
    for d in matrix:
        keys = sorted([key for key in d])
        
    for key in keys:
        print("{}: ".format(key), end="")
        for i in range(len(matrix)):
            print("{}".format(matrix[i][key]), end=" ")
        print("")
        
    return matrix
```

```python
consensus_and_profile("rosalind_cons.txt")

```
