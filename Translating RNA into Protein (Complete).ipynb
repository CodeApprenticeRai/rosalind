```python
import pandas as pd
import csv

rdata0 = []

#read file
with open('rna_codon_table','r') as f:
    reader=csv.reader(f,delimiter='\t')
    for thing in reader:
        rdata0.append(thing)
        
#data = pd.read_csv("rna_codon_table", sep="\t")

pairs = []


#break into substrings 1
for arr in rdata0:
    for s in arr:
        pairs.append(s.split("      "))

rdata1 = []

#break into substrings 2
for arr in pairs:
    for substring_pair in arr:
        rdata1.append(substring_pair)

#uneccessary lengthy filtering
req = len(rdata1[0])
       
outliers0 = [ rdata1[i] for i in range(len(rdata1)) if len(rdata1[i]) > req ]
outliers1 = [s.split("   ") for s in outliers0] 
outliers = []
rdata2 = []

for i in range(0,len(rdata1)):    
    n =len(rdata1[i])
    if  n <= req:
        rdata2.append(rdata1[i])

for arr in outliers1:
    for substring_pair in arr:
        outliers.append(substring_pair)


protein_string = 'AUGGCCAUGGCGCCCAGAACUGAGAUCAAUAGUACCCGUAUUAACGGGUGA'
sub_strings = [protein_string[i:i+3] for i in range(0,len(protein_string),3)]
sub_strings

rdata = rdata2 + outliers

#Data ready to be added to turn into a dict

rdata

for string in rdata:
    final_sub_string = string.split(' ')
    data[final_sub_string[0]] = final_sub_string[1]



def translate(protein_string, trans_key):
    sub_strings = [protein_string[i:i+3] for i in range(0,len(protein_string),3)]
    
    answer = ''
    for ss in sub_strings:
        answer += trans_key[ss]
    return answer

translate(protein_string, data)
```
